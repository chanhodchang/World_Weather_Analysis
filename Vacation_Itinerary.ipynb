```python
import pandas as pd
import gmaps
import requests
from citipy import citipy
from config import g_key
import numpy
from datetime import datetime
import time
from scipy.stats import linregress
```

```python
city_data=pd.read_csv('challenge_weather_data/WeatherPy_vacation.csv', index_col=0)
city_data.head()
```

```python
au_cities=city_data.loc[(city_data['Country']=='AU')]
au_cities
```

```python
lithgow=au_cities.loc[au_cities['City']=='Lithgow']
nelson_bay=au_cities.loc[au_cities['City']=='Nelson Bay']
port_macquarie=au_cities.loc[au_cities['City']=='Port Macquarie']
nambucca_heads=au_cities.loc[au_cities['City']=='Nambucca Heads']

lithgow
```

```python
au_vac_cities=au_cities.loc[au_cities['City'].isin(['Lithgow','Nelson Bay','Port Macquarie','Nambucca Heads'])]
au_vac_cities
```

```python
import gmaps.datasets
gmaps.configure(api_key=g_key)
```

```python
lithgow_t=(-33.48,150.16)
nelson_bay_t=(-32.72,152.14)
port_macquarie_t=(-31.43,152.91)
nambucca_heads_t=(-30.64,152.99)

fig=gmaps.figure()
au_map=gmaps.directions_layer(lithgow_t, nambucca_heads_t, waypoints=[nelson_bay_t,port_macquarie_t], travel_mode='DRIVING')
fig.add_layer(au_map)
fig
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
hotel_info=[info_box_template.format(**row) for index, row in au_vac_cities.iterrows()]
```

```python
locations=au_vac_cities[['Lat','Lng']]
fig=gmaps.figure(center=(30.0,31.0), zoom_level=1.5)
marker_layer=gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig
```
